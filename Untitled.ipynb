```python
import numpy as np
import matplotlib.pyplot as plt
from few.waveform import FastSchwarzschildEccentricFlux

# Initialize the waveform model
wave = FastSchwarzschildEccentricFlux()

# Parameters
M    = 1e6   # MBH mass (solar masses)
mu   = 10.0  # secondary mass (solar masses)
a    = 0.0   # spin (0 for Schwarzschild)
p0   = 12.0  # semi-latus rectum
e0   = 0.4   # eccentricity
Y0   = 1.0   # cos(inclination) — equatorial

dist  = 1.0  # Gpc
qS, phiS = 0.5, 0.3
qK, phiK = 0.8, 0.1

T  = 1.0   # years
dt = 10.0  # seconds

# Generate waveform
h = wave(M, mu, p0, e0,
         qS, phiS,
         dist,
         Phi_phi0=0.0, Phi_r0=0.0,
         T=T, dt=dt)

print(f"Samples: {len(h)}")
print(f"Duration: {len(h)*dt/3600/24:.1f} days")
print(f"Max strain: {np.max(np.abs(h)):.2e}")
```

```python
# Time array in days
t = np.arange(len(h)) * dt / 3600 / 24

fig, axes = plt.subplots(3, 1, figsize=(12, 8))

# Full inspiral - h+
axes[0].plot(t, h.real, lw=0.3, color='steelblue')
axes[0].set_ylabel('h+')
axes[0].set_title('Full 1-year EMRI inspiral')

# Zoom in on the first few days
mask = t < 5
axes[1].plot(t[mask], h.real[mask], lw=0.8, color='steelblue')
axes[1].set_ylabel('h+')
axes[1].set_title('First 5 days (orbital structure visible)')

# Zoom in near plunge (last few days)
mask2 = t > 360
axes[2].plot(t[mask2], h.real[mask2], lw=0.8, color='tomato')
axes[2].set_ylabel('h+')
axes[2].set_title('Final days before plunge (frequency runup)')

for ax in axes:
    ax.set_xlabel('Time (days)')

plt.tight_layout()
plt.show()
```

```python
# FFT
N = len(h)
freqs = np.fft.rfftfreq(N, d=dt)  # Hz
H = np.fft.rfft(h.real)
power = np.abs(H)**2

# Convert to mHz
freqs_mHz = freqs * 1e3

fig, ax = plt.subplots(figsize=(12, 4))
ax.semilogy(freqs_mHz, power, lw=0.5, color='steelblue')
ax.set_xlim(0, 10)
ax.set_xlabel('Frequency (mHz)')
ax.set_ylabel('Power')
ax.set_title('EMRI power spectrum — comb structure')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
from scipy.signal import spectrogram
from skimage.transform import radon
import matplotlib.pyplot as plt
import numpy as np

# Compute spectrogram
f, t_spec, Sxx = spectrogram(h.real, fs=1/dt, 
                              nperseg=8192, 
                              noverlap=4096)

# Focus on LISA band (0.1 - 10 mHz)
f_mHz = f * 1e3
mask = (f_mHz > 0.1) & (f_mHz < 10)

Sxx_band = Sxx[mask]
f_band = f_mHz[mask]

print(f"Spectrogram shape: {Sxx_band.shape}")
print(f"Frequency range: {f_band[0]:.3f} - {f_band[-1]:.3f} mHz")
print(f"Time bins: {len(t_spec)}")
```

```python
plt.figure(figsize=(12, 5))
plt.pcolormesh(t_spec / 3600 / 24, f_band, np.log10(Sxx_band + 1e-30), 
               shading='auto', cmap='inferno')
plt.colorbar(label='log10 Power')
plt.xlabel('Time (days)')
plt.ylabel('Frequency (mHz)')
plt.title('EMRI Spectrogram — chirping harmonic tracks')
plt.tight_layout()
plt.show()
```

```python
from skimage.transform import radon

# Normalize the spectrogram to [0,1] for Radon
img = np.log10(Sxx_band + 1e-30)
img = (img - img.min()) / (img.max() - img.min())

# Radon transform — angles near 90° = nearly horizontal lines
# Our tracks are nearly horizontal so we focus there
angles = np.linspace(80, 100, 200)
sinogram = radon(img, theta=angles)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Original spectrogram
axes[0].imshow(img, aspect='auto', origin='lower', cmap='inferno',
               extent=[t_spec[0]/3600/24, t_spec[-1]/3600/24,
                       f_band[0], f_band[-1]])
axes[0].set_xlabel('Time (days)')
axes[0].set_ylabel('Frequency (mHz)')
axes[0].set_title('Spectrogram')

# Radon transform (sinogram)
axes[1].imshow(sinogram, aspect='auto', origin='lower', cmap='inferno',
               extent=[angles[0], angles[-1], 0, sinogram.shape[0]])
axes[1].set_xlabel('Angle (degrees)')
axes[1].set_ylabel('Radon coordinate')
axes[1].set_title('Radon Transform — each track becomes a peak')

plt.tight_layout()
plt.show()
```

```python
# Isolate a narrow frequency band around one strong harmonic
# pick the ~2.5 mHz track which looks bright
f_lo, f_hi = 2.3, 2.7
band_mask = (f_band > f_lo) & (f_band < f_hi)

img_single = np.log10(Sxx_band[band_mask] + 1e-30)
img_single = (img_single - img_single.min()) / (img_single.max() - img_single.min())

angles = np.linspace(85, 95, 500)
sinogram_single = radon(img_single, theta=angles)

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].imshow(img_single, aspect='auto', origin='lower', cmap='inferno',
               extent=[t_spec[0]/3600/24, t_spec[-1]/3600/24, f_lo, f_hi])
axes[0].set_xlabel('Time (days)')
axes[0].set_ylabel('Frequency (mHz)')
axes[0].set_title('Single harmonic track')

axes[1].imshow(sinogram_single, aspect='auto', origin='lower', cmap='inferno',
               extent=[angles[0], angles[-1], 0, sinogram_single.shape[0]])
axes[1].set_xlabel('Angle (degrees)')
axes[1].set_ylabel('Radon coordinate')
axes[1].set_title('Radon of single track — one clean peak')

plt.tight_layout()
plt.show()
```

```python
f_lo, f_hi = 2.48, 2.52  # tight window around one track
band_mask = (f_band > f_lo) & (f_band < f_hi)

img_single = np.log10(Sxx_band[band_mask] + 1e-30)
img_single = (img_single - img_single.min()) / (img_single.max() - img_single.min())

angles = np.linspace(85, 95, 500)
sinogram_single = radon(img_single, theta=angles)

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

axes[0].imshow(img_single, aspect='auto', origin='lower', cmap='inferno',
               extent=[t_spec[0]/3600/24, t_spec[-1]/3600/24, f_lo, f_hi])
axes[0].set_xlabel('Time (days)')
axes[0].set_ylabel('Frequency (mHz)')
axes[0].set_title('Tight single track window')

axes[1].imshow(sinogram_single, aspect='auto', origin='lower', cmap='inferno',
               extent=[angles[0], angles[-1], 0, sinogram_single.shape[0]])
axes[1].set_xlabel('Angle (degrees)')
axes[1].set_ylabel('Radon coordinate')
axes[1].set_title('Radon — should be one localized peak')

plt.tight_layout()
plt.show()
```

```python
# Full spectrogram Radon, find peak angle precisely
img_full = np.log10(Sxx_band + 1e-30)
img_full = (img_full - img_full.min()) / (img_full.max() - img_full.min())

angles = np.linspace(80, 100, 500)
sinogram_full = radon(img_full, theta=angles)

# Find the angle with maximum total power
radon_power = sinogram_full.sum(axis=0)

peak_angle = angles[np.argmax(radon_power)]
print(f"Peak Radon angle: {peak_angle:.3f} degrees")

# Convert angle to chirp rate (df/dt)
# angle=90 means horizontal (zero chirp), deviation encodes df/dt
df_total = f_band[-1] - f_band[0]   # mHz
dt_total = t_spec[-1] - t_spec[0]   # seconds
chirp_rate = np.tan(np.radians(90 - peak_angle)) * (df_total / dt_total) * 1e-3  # Hz/s
print(f"Estimated chirp rate: {chirp_rate:.3e} Hz/s")

plt.figure(figsize=(10, 3))
plt.plot(angles, radon_power)
plt.axvline(peak_angle, color='red', linestyle='--', label=f'Peak: {peak_angle:.2f}°')
plt.xlabel('Angle (degrees)')
plt.ylabel('Integrated Radon power')
plt.title('Radon power vs angle — peak encodes chirp rate')
plt.legend()
plt.tight_layout()
plt.show()
```
